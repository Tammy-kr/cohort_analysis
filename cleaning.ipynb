```python
#import txt file and convert txt to csv
import pandas as pd
import csv
filenametxt = "C:/Users/Tammy/data/test/testset.txt"
filenamecsv = "C:/Users/Tammy/data/test/testsetnew.csv"
with open(filenametxt, 'r') as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split(",") for line in stripped if line)
    with open(filenamecsv, 'w') as out_file:
        writer = csv.writer(out_file, delimiter=",")
        writer.writerow(())
        writer.writerows(lines)
```

```python
# cretate correct names for columns
col_names = ['Created Date', 'Player ID', 'Payment Method', 'System Currency Amount', 'Number Deposit', 'Tx Direction In\Out', 'Country → Name Short En', 'Player → Registration Date']
df = pd.read_csv('testsetnew.csv', names=col_names)
```

```python
df.head(20)
```

```python
#drop extra rows
data_drop1 = df.drop(labels=[0,1,2], axis=0)
```

```python
data_drop1.head(5)
```

```python
#transform data

import numpy as np

X = data_drop1.iloc[:, :].values
Y = data_drop1.iloc[:, 0].values
Y1 = data_drop1.iloc[:,1].values

X = X.tolist()
Y = Y.tolist()
Y1 = Y1.tolist()
Z = []

for i in range(0,len(data_drop1),2):
    L = X[i]
    L.append(Y[i])
    L.append(Y[i+1])
    L.append(Y1[i])
    L.append(Y1[i+1])
    Z.append(L)

Z = np.array(Z, dtype=object)
df_new = pd.DataFrame(Z,index=None)
df_new.to_csv("test_new_drop1.csv")
```
